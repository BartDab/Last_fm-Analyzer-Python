```python
print(df['Date'])
df['Date']=pd.to_datetime(df['Date'], format='%d %b %Y %H:%M')
print(df['Date'])
```

```python
#################
### ABANDONED ###
#################

start_date1 = pd.to_datetime('2018-03-25 02:00:00')
end_date1 = pd.to_datetime('2018-10-28 03:00:00')
start_date2 = pd.to_datetime('2019-03-31 02:00:00')
end_date2 = pd.to_datetime('2019-10-27 03:00:00')
start_date3 = pd.to_datetime('2020-03-29 02:00:00')
end_date3 = pd.to_datetime('2017-10-25 03:00:00')
filtered=df[((df['Date']>start_date1)&(df['Date']<end_date1)) | ((df['Date']>=start_date2) & (df['Date']<end_date2)) | ((df['Date']>=start_date3) & (df['Date']<end_date3))]
df['DST']=((df['Date']>start_date1)&(df['Date']<end_date1)) | ((df['Date']>=start_date2) & (df['Date']<end_date2)) | ((df['Date']>=start_date3) & (df['Date']<end_date3))
dst_correction = pd.DateOffset(1)
print(df[df['DST']==1].head())
for record in df:
    if df['DST'][record]==True:
        df['Date'][record]+=dst_correction
print(df[df['DST']==1].head())
```

```python
monthly_scrobbles=df['Date'].groupby([df.Date.dt.year, df.Date.dt.month]).agg('count')
monthly_scrobbles
fig, ax = plt.subplots()
ax = monthly_scrobbles.plot(kind='bar', figsize=[10,5],linewidth=4,alpha=0.8,color='yellowgreen')
ax.set_xlabel(' ')
plt.show()
```

```python
daily_scrobbles=df['Date'].groupby([df.Date.dt.year, df.Date.dt.month, df.Date.dt.day]).agg('count')
fig, ax = plt.subplots()
ax = daily_scrobbles.plot(kind='line', figsize=[10,5],linewidth=1,alpha=0.8,color='yellowgreen')
ax.set_xlabel(' ')
plt.show()
```
